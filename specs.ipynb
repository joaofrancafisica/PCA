```python
%matplotlib inline
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,50,61)
y = np.linspace(3000,9000,601)
```

```python
surf = pd.read_csv("SN2012fr.dat", sep = '\s+', header = None)
```

```python
spec = pd.Series(surf[0])
```

```python
spec = np.reshape(np.array(spec), (601, 61)).T
```

```python
N = 11
plotspecs = np.linspace(0, 60, N, dtype = int)
plotspecs = plotspecs[::-1]
```

```python
offset = 0
plt.figure(figsize=(6, 8))
for i in range(N):
    
    plt.plot(y, spec[plotspecs[i]]/max(spec[plotspecs[i]]) + offset, 'k')
    
    offset += 0.7
    
plt.yticks([])
plt.xlabel("wavelength ($\AA$)")
plt.ylabel("Flux + offset")
plt.xlim(3500, 8000)
plt.savefig("specs2012fr.jpg")   # save the figure to file
plt.show()
```
